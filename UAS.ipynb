```python
import pandas as pd
import matplotlib.pyplot as plt
```

Studi Kasus 1 = Produk dengan harga termurah

Dataset: Produk

```python
csv_path = './produk.csv'
df = pd.read_csv(csv_path)
df_sorted = df.sort_values('Price', ascending=True)

df_sorted = df_sorted[df_sorted['Price'] != 0]
df_sorted.head()
```

```python
x = df_sorted['id'].astype(str).head(10)
y = df_sorted['productName'].head(10)
z = df_sorted['Price'].head(10)


plt.scatter(x, y, c=z)
plt.xlabel('ID')
plt.ylabel('Product')
plt.title('10 Produk Termurah')
plt.colorbar(label='Price')
plt.show()
```

Studi Kasus 2 = Angka kebutuhan Provinsi Jawa Barat Per Tahun

Dataset: Kebutuhan_per_bulan_2

```python
csv_path = './kebutuhan_per_bulan_2.csv'
df = pd.read_csv(csv_path)
df_prov = df[df['Propinsi'] == "JAWA BARAT"]
df_sorted_prov = df_prov.sort_values('tahun', ascending=True)
df_sorted_prov.head(11)
```

```python
x = df_sorted_prov['tahun']
y = df_sorted_prov['kebutuhan']

plt.scatter(x, y)
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')
plt.title('Angka kebutuhan Jawa Barat Per Tahun')
plt.show()
```

Studi Kasus 3 = Penyebaran pelanggan

Studi Kasus 4 = Penjualan Perbulan

Dataset Supply Kantor

```python
csv_path = './SupplyKantor.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
df["total"]= df["Jumlah"] * df["Harga Satuan"]

df['TanggalPesan'] = pd.to_datetime(df['TanggalPesan'], format='%d-%b-%y')
df['month'] = df['TanggalPesan'].dt.month

df.sort_values('month', ascending = True)
x = df['TanggalPesan']
y = df['total']

plt.scatter(x, y)
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Total Pemasukan')
plt.title('Penjualan Perbulan')
plt.show()
```
